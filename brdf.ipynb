# Light Scattering

When light hits on a surface, the behaviors vary depending on the properties of the surface. Light can be **absorbed, reflected, and transmitted**.

In rendering we typically only consider the reflected and transmitted light, both phenomena can be referred as scattering.

### Ideal diffuse reflection:

light is reflected in all directions

colored by surface color

### Ideal specular reflection:

light reflected/refracted in single direction respectively.

colored by source color.

We further divide objects into two groups: opaque and translucent.





opaque objects exhibit little or no transmittance, while translucent objects can have reflectance and transmittance.




A few simplied analytical models are developed:

such as Blinn-Phong(plastic), Cook-Torrance(better specular), Ward(support anisotropic), Oren-Nayar.







We model reflection as mainly mixture of a few "ideal" reflection functions:

1. Ideal Diffuse: This is the simplest form of reflection. which means light will be **uniformly** scattered on the upper hemisphere of the light-surface intersection point.

2. Ideal Specular: When the incoming light is reflected perfectly on the mirror direction around the normal vector of the surface. 

3. Ideal Specular Transmission: 

Most material models will be a combined function of the two.
